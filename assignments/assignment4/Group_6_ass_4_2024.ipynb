### 2. Reshape the dataset from wide to long at year level. Drop columns if they are in Dollars. Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, and the name of the firm.


```python
# Filter the columns that meet the criteria.
filtered_columns = ['Company name Latin alphabet']

# Find the columns that starts with "Operating Revenue (Turnover)\nth LCU"
revenue_columns = [col for col in df.columns if col.startswith('Operating revenue (Turnover)\nth LCU')]

# Add the selected columns to the list.
filtered_columns.extend(revenue_columns)

# Filter the DataFrame.
df_filtered = df[filtered_columns]
```

```python
df_filtered
```

```python
# Use the melt function to reshape the DataFrame
df_long = pd.melt(df_filtered, id_vars=['Company name Latin alphabet'], var_name='Year', value_name='Operating Revenue (Turnover)')
df_long
```

```python
# Extract the year from each column name into 'Year'.
df_long['Year'] = df_long['Year'].str.extract('(\d{4})')

# Rename the original columns that are in the 'Year' column.
df_long = df_long.rename(columns=lambda x: f'Revenue_{x}' if x.isnumeric() else x)

# Print the resulting DataFrame.
df_long
```

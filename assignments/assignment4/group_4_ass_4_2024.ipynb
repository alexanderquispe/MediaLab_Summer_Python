```python
def calling_data(firm_name, start_year, end_year=None):
    # Filtering data based on firm_name and start_year
    filtered_data = data_frame[(data_frame['Company name Latin alphabet'].isin(firm_name)) & (data_frame['Last avail. year'] == start_year)]
```

```python
# If end_year is provided, filter data for the specified range
    if end_year:
        filtered_data = filtered_data[(filtered_data['Last avail. year'] >= start_year) & (filtered_data['Last avail. year'] <= end_year)]
```

## 2. Reshape the dataset from wide to long at year level. Drop columns if they are in Dollars. Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, and the name of the firm. 

First we have to eliminate the dollar columns, to achieve this first we create a for loop that allows us to select the columns that hace USD so then we can drop the using the Drop function.

```python
#We safe the selection in the object columns_to_drop
columns_to_drop = []
for column in DataFrame.columns:
    if 'USD' in column:
        columns_to_drop.append(column)
#We use the function drop
DataFrame.drop(columns=columns_to_drop, inplace=True)
#We corroborate our progress
DataFrame
```

Then we use the melt function, in our id_vars we put the name of the column where the firms name are. Then we designate all the operating revenues in the value_vars. Then we put the new columns name that are year and the Operating revenue LCU.

```python
Long_df = pd.melt(DataFrame, id_vars=['Company name Latin alphabet'],
                    value_vars=DataFrame.columns[DataFrame.columns.str.contains('LCU')],
                    var_name='Year', value_name='Operating revenue (Turnover) th LCU')
Long_df
```

Finally we modify our year column so we only see the year. For that we extract the year

```python
Long_df['Year'] = Long_df['Year'].astype(str).str.extract('(\d+)').astype(int)
Long_df
```

5. Generate a function named as calling_data that take as input the `name of the firm  or list of names (firm_name)` and `(start_year, end_year)`. The function should return a tuple with a dataframe with the same format as the one resulting from task 1-4, and your dictionary of columns. The function should work without `end_year`. It means it only takes `start_year`.

We create the function where we put two conditionals. In both we use isintance() function that allows us to find if objetcs are of a certain type. Our first conditional identifies if firm_name is a string or a list to then select the rows that have the firm_name in the Firm column. The second conditional is double. First it selects rows from year_star to year_end only if end_year an start_yeat are intergers. Then if start_year is an interger and end_year is none (it's not defined) then it only selects rows that have start_year. Otherwise it would print an error. 

```python
def calling_data(firm_name, start_year, end_year):
    #We use isintance to use conditional in case firm_name is one or more comp
    if isinstance(firm_name, str):
        selected_rows = Long_df[Long_df['Firm'] == firm_name]
    elif isinstance(firm_name, list):
        selected_rows = Long_df[Long_df['Firm'].isin(firm_name)]
    else:
        raise ValueError("firm_name should be a string or a list of strings.")

    if isinstance(end_year, int) and isinstance(start_year, int):
        selected_rows = selected_rows[(selected_rows['Year'] >= start_year) & (selected_rows['Year'] <= end_year)]
    elif end_year is None and isinstance(start_year, int):
        selected_rows = selected_rows[selected_rows['Year'] == start_year]
    else:
        raise ValueError("start_year and end_year should be integers.")

    the_tuple = (selected_rows, dictionary)

    return the_tuple
```

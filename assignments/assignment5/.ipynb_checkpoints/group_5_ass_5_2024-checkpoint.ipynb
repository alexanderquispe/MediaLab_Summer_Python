```python
# Clean Table
import unidecode
import pandas as pd

# Assuming table is a DataFrame
# Clean Table
row_new_columns = table[0].iloc[0, 2:]
clean_columns = row_new_columns \
    .str.replace(" ", "_") \
    .str.lower().str.replace("%", "share_") \
    .apply(lambda x: unidecode.unidecode(x) if isinstance(x, str) else x) \
    .tolist()

# Selecting specific columns
table_clean = table[0].iloc[1:, 2:].copy()
table_clean 
```

```python
# Store all_tables
all_tables = {}

# Iterate all 26 departments
for dpt_idx in range(26):
    try:
        
        # Click on the department drop-down list
        dpt = Select(driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[2]/select"))
        
        # Get department based on iterator
        department = dpt.options[dpt_idx]

        # Get the department name
        dpt_name = department.text

        # Avoid selecting the "--TODOS--" option
        if dpt_name != "--TODOS--":
            
            # Click on the department
            department.click()

            # Get all provinces of the selected department
            prov = Select(driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[3]/select"))
            
            # We use the len function to obtain the number of provinces
            num_prov_options = len(prov.options)

            # Run all the provinces
            for prov_idx in range(num_prov_options):
                try:
                    
                    # Call back the HTML reference of the updated dropdown list
                    prov = Select(driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[3]/select"))
                    
                    # Get the province according to the iterator
                    province = prov.options[prov_idx]

                    # Get the name of the province
                    prov_name = province.text
                    
                    # Avoid selecting the "--TODOS--" option from the province drop-down list
                    if prov_name != "--TODOS--":
                        
                        # Click on the province
                        province.click()

                        # Get all the provinces of the selected province
                        dist = Select(driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[4]/select"))
                        
                        # We use the len function to obtain the number of districts
                        num_dist_options = len(dist.options)

                        # Run all the districts
                        for dist_idx in range(num_dist_options):
                            
                            try:
                                
                                # Call back the HTML reference of the updated dropdown list
                                dist = Select(driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[4]/select"))
                                
                                # Get the district based on the iterator
                                district = dist.options[dist_idx]

                                # Get district name
                                dist_name = district.text

                                # Avoid selecting the "--SELECCIONE--" option from the district drop-down list
                                if dist_name != "-- SELECCIONE --":
                                    # Click on district
                                    district.click()

                                    # Use split to convert the URL to a list. Also, access the last element to obtain the ubigeo
                                    ubigeo = driver.current_url.split("/")[ -1 ]

                                    # Get the reference to the table
                                    table_path = driver.find_element(By.XPATH, "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[4]/div[1]/div[3]/div" )
                                    table_html = table_path.get_attribute( 'innerHTML' )
                                    
                                    # Use pandas to read table
                                    table = pd.read_html( table_html )

                                    # Clean data
                                    # Replace empty spaces with underscore
                                    # Convert text to lowercase and replace % with share_
                                    # Convert Unicode characters
                                    row_new_columns = table[ 0 ].iloc[ 0 , 2: ]
                                    clean_columns = row_new_columns \
                                                          .str.replace(" ","_") \
                                                          .str.lower().str.replace( "%", "share_") \
                                                          .apply( lambda x : unidecode.unidecode( x ) ) \
                                                          .tolist()

                                    # Select columns
                                    table_clean = table[0].iloc[ 1:, 2: ].copy()

                                    # Rename columns
                                    table_clean.columns = clean_columns

                                    # Define values to the columns
                                    table_clean[ 'department' ] = dpt_name
                                    table_clean[ 'province' ]   = prov_name
                                    table_clean[ 'district' ]   = dist_name


                                    # store tables
                                    all_tables[ ubigeo ] = table_clean

                            except Exception as dist_error:
                                print(f"Error in district loop: {dist_error}")
                                # Catch errors in the district loop

                except Exception as prov_error:
                    print(f"Error in province loop: {prov_error}")
                    # Catch errors in the province loop

    except Exception as dpt_error:
        print(f"Error in department loop: {dpt_error}")
        # Manage department loop exception
```

```python
# Concatenate the values from the "all_tables" dictionary and reset the index
final_data = pd.concat( all_tables.values() ).reset_index( drop = True ) 
```

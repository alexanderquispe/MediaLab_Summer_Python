###5.After reshaping the data, you should replicate the long data, but instead of using methods of pandas, you have to use for loop.

```python
#Crea nuevo objeto
ejercicio_5 = pd.DataFrame()
#Se escoge 'CompanyName' como la columna definida
empresas_unicas = data_frame['CompanyName'].unique()
#Iteración
for empresa in empresas_unicas:
    empresa_data = data_frame[data_frame['CompanyName'] == empresa]
ejercicio_5 = pd.concat([ejercicio_5, empresa_data])
ejercicio_5.reset_index(drop=True, inplace=True)
print(ejercicio_5)
```

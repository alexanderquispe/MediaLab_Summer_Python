```python
################################################################   EXERCISE 2    #####################

########### PARTE1


# Specific characters to look for
specific_chars = 'Company '
specific_chars_2 ='Operating revenue'

# Identify columns that contain the specific characters
columns_to_keep_1 = [column for column in data.columns if specific_chars  in column]
columns_to_keep_2 = [column for column in data.columns if specific_chars_2 in column]


# Drop these columns
columns_to_keep_3 = columns_to_keep_1 +  columns_to_keep_2

# Select only the columns to keep
data = data[columns_to_keep_3]
# The modified DataFrame
data.head()
```

```python
########### PART 2

# Specific characters to look for
specific_chars = 'Company '
specific_chars_2 ='Operating revenue'

# Identify columns that contain the specific characters
columns_to_keep_1 = [column for column in data.columns if specific_chars  in column]
columns_to_keep_2 = [column for column in data.columns if specific_chars_2 in column]


# Drop these columns
columns_to_keep_3 = columns_to_keep_1 +  columns_to_keep_2

# Select only the columns to keep
data = data[columns_to_keep_3]
# The modified DataFrame
data.head()
```

```python
########################################## Reshaping  ##########################################


# Reshaping the dataset from wide to long format at the year level
# 'id_vars' is used to keep the 'Company name Latin alphabet' column as is
# 'var_name' names the new column that will contain the years
# 'value_name' names the new column that will contain the operating revenues
reshaped_df = data.melt(id_vars=["Company name Latin alphabet"], var_name="Year", value_name="Operating Revenue LCU")

# Extracting the year from the 'Year' column
# The lambda function takes each entry in 'Year', splits it by spaces, and takes the last part (which is the year)
reshaped_df['Year'] = reshaped_df['Year'].apply(lambda x: x.split()[-1])

reshaped_df.head()  # Displaying the first 5 rows of the reshaped dataframe
```

```python
########################################## EXCERCISE 3 ##########################################
# Renaming the first column to 'Company name'
data.rename(columns={data.columns[0]: 'Company name'}, inplace=True)

# Renaming the other columns to only the last 8 characters
data.columns = [col if i == 0 else col[-8:] for i, col in enumerate(data.columns)]

data.head() 
```

```python
########################################## EXCERCISE 4 ##########################################



# Generating a dictionary with the definition of all the columns in the filtered_df dataset

column_definitions = {
    "Company name Latin alphabet": "Name of the company in Latin alphabet",
}

# Looping through the remaining columns, which represent years
for col in data.columns[1:]:
    year = col.split()[-1]
    column_definitions[col] = f"Operating revenue in local currency unit (LCU) for the year {year}"

column_definitions
```

```python
########################################## EXCERCISE 5 ##########################################


# Replicating the reshaping of the data using a for loop instead of pandas methods

# Initialize an empty list to store the reshaped data
reshaped_data = []

# Loop over each row in the filtered dataframe
for index, row in data.iterrows():
    company_name = row["Company name"]

    # Loop over each year column
    for year_col in data.columns[1:]:
        year = year_col.split()[-1]  # Extract the year from the column name
        revenue = row[year_col]      # Get the revenue for that year

        # Append a dictionary for each year with company name, year, and revenue
        reshaped_data.append({"Company name Latin alphabet": company_name, "Year": year, "Operating Revenue LCU": revenue})

# Convert the list of dictionaries to a DataFrame
reshaped_df_for_loop = pd.DataFrame(reshaped_data)

reshaped_df_for_loop.head()  # Displaying the first 5 rows of the new reshaped dataframe
```

```python
########################################## EXCERCISE 6 ##########################################


# Replicating the reshaping of the data using a while loop instead of a for loop

# Initialize an empty list to store the reshaped data
reshaped_data_while = []

row_index = 0
num_rows = len(data)

# Loop over each row in the filtered dataframe using a while loop
while row_index < num_rows:
    row = data.iloc[row_index]
    company_name = row["Company name"]

    col_index = 1  # Starting from the first year column
    num_cols = len(data.columns)

    # Loop over each year column
    while col_index < num_cols:
        year_col = data.columns[col_index]
        year = year_col.split()[-1]  # Extract the year from the column name
        revenue = row[year_col]      # Get the revenue for that year

        # Append a dictionary for each year with company name, year, and revenue
        reshaped_data_while.append({"Company name Latin alphabet": company_name, "Year": year, "Operating Revenue LCU": revenue})

        col_index += 1

    row_index += 1

# Convert the list of dictionaries to a DataFrame
reshaped_df_while_loop = pd.DataFrame(reshaped_data_while)

reshaped_df_while_loop.head()  # Displaying the first 5 rows of the new reshaped dataframe
```

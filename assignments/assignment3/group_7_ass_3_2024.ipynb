# 2. Reshape the dataset from wide to long at year level. Drop columns if they are in Dollars. Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, and the name of the firm.

```python
columns_dollars = [col for col in data.columns if 'USD' in col]
without_dollars = data.drop(columns=columns_dollars)
```

```python
tokeep = [col for col in without_dollars.columns if col == 'Operating revenue (Turnover)\nth LCU 2023' or col == 'Company name Latin alphabet']
selected = without_dollars[tokeep]
```

```python
long = selected.melt(id_vars=['Company name Latin alphabet'], var_name='Year', value_name='LCU')
```

```python
long['Year'] = long['Year'].str.extract('(\d{4})') 
```
